```python
# Install required libraries
!pip install transformers

# Import necessary libraries
from transformers import AutoModelForSequenceClassification, AutoTokenizer
import torch
```

```python
# Define the model name
model_name = "MSarah/ComplainClassificationFineTunedBert"

# Load the model and tokenizer from the Hugging Face Hub
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Define a function to make predictions
def predict(text, model, tokenizer):
    # Tokenize the input text
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)

    # Forward pass, get logits
    with torch.no_grad():
        outputs = model(**inputs)
    logits = outputs.logits

    # Get the predicted class
    predicted_class = torch.argmax(logits, dim=1).item()

    if predicted_class == 0:
      result = 'Account Services'
    elif predicted_class == 1:
      result ='Others'
    elif predicted_class == 2:
      result ='Mortgage/Loan'
    elif predicted_class == 3:
      result ='Credit card or prepaid card'
    else:
      result = 'Theft/Dispute Reporting'

    return result

# Sample data to test the model
sample_texts = [
    "My loan is not processed yet",
    "I want to open an account",
    "I want to extend the limit of my credit card",
    "Please share timings of the your branches."
]

# Make predictions on sample data
for text in sample_texts:
    predicted_class = predict(text, model, tokenizer)
    print(f"Text: {text}\nPredicted class: {predicted_class}\n")
```
